```python
import numpy as np
import pandas as pd
```

```python
matches = pd.read_csv('matches.csv')
matches.head()
```

```python
def Bowlingteam(row):
    if (row['Team1'] == row['TossWinner']):
        if row['TossDecision'] == 'bat':
            return row['Team1']
        else:
            return row['Team2']
    else:
        if row['TossDecision'] == 'bat':
            return row['Team2']
        else:
            return row['Team1']
```

```python
matches['BowlingTeam'] = matches.apply(Bowlingteam,axis=1)
```

```python
matches
```

```python
delivery = pd.read_csv('deliveries.csv')
delivery.head()
```

```python
total_score_df = delivery.groupby(['ID','innings']).sum()['total_run'].reset_index()
```

```python
total_score_df
```

```python
total_score_df = total_score_df[total_score_df['innings']==1]
```

```python
total_score_df
```

```python
match_df = matches.merge(total_score_df[['ID','total_run']],left_on='ID',right_on='ID')
```

```python
match_df.head()
```

```python
match_df['Team1'].unique()
```

```python
teams = [
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Chennai Super Kings',
    'Kolkata Knight Riders',
    'Punjab Kings',
    'Delhi Capitals',
    'Gujarat Titans',
    'Lucknow Super Giants',
    'Rajasthan Royals'
]
```

```python
match_df['Team1'] = match_df['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['Team2'] = match_df['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['Team1'] = match_df['Team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['Team2'] = match_df['Team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

```

```python
match_df['Team1'].unique()
```

```python
match_df['Team1'] = match_df['Team1'][match_df['Team1'].isin(teams)]
match_df['Team2'] = match_df['Team2'][match_df['Team2'].isin(teams)]
```

```python
match_df['Team1'].unique()
```

```python
match_df.shape
```

```python
match_df
```

```python
match_df = match_df[match_df['method'].isnull()]
```

```python
match_df.shape
```

```python
match_df = match_df[['ID','Venue','WinningTeam','BowlingTeam','total_run']]
```

```python
match_df.head()
```

```python
delivery_df = match_df.merge(delivery,on='ID')
```

```python
delivery_df.head()
```

```python
delivery_df = delivery_df[delivery_df['innings']== 2]
```

```python
delivery_df.shape
```

```python
delivery_df['total_run_y'] = pd.to_numeric(delivery_df['total_run_y'], errors='coerce')
delivery_df['current_score'] = delivery_df.groupby('ID')['total_run_y'].cumsum()
```

```python
delivery_df
```

```python
delivery_df['runs_left'] = delivery_df['total_run_x']-delivery_df['current_score']
```

```python
delivery_df
```

```python
delivery_df['balls_left'] = 120 - (delivery_df['overs']*6 + delivery_df['ballnumber'])
```

```python
delivery_df
```

```python
delivery_df['wickets_left'] = 10 - delivery_df.groupby('ID')['isWicketDelivery'].cumsum()
```

```python
delivery_df['wickets_left'].unique()
```

```python
delivery_df['crr'] = (delivery_df['current_score']*6)/(120-delivery_df['balls_left'])
```

```python
delivery_df
```

```python
delivery_df['rrr'] = (delivery_df['runs_left']*6)/(delivery_df['balls_left'])
```

```python
delivery_df
```

```python
def result(row):
    return 1 if row['BattingTeam'] == row['WinningTeam'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
final_df = delivery_df[['BattingTeam','BowlingTeam','Venue','runs_left','balls_left','wickets_left','total_run_x','crr','rrr','result']]
```

```python
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left']!=0]
```

```python
X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([('trf',OneHotEncoder(sparse=False,drop='first'),['BattingTeam','BowlingTeam','Venue'])
                        ]
                        ,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
pipe = Pipeline(steps=[
    ('step1', trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(X_test)[10]
```

```python
def match_progression(x_df,match_id,pipe):
    match = x_df[x_df['ID'] == match_id]
    match = match[(match['ballnumber'] == 6)]
    temp_df = match[['BattingTeam','BowlingTeam','Venue','runs_left','balls_left','wickets_left','total_run_x','crr','rrr']].dropna()
    temp_df = temp_df[temp_df['balls_left'] != 0]
    if temp_df.empty:
        print("DataFrame is empty. Unable to make predictions.")
        return None, None
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
    
    target = temp_df['total_run_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets_left'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
    
    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
```
